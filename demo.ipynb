A demo notebook to reproduce the qualitative examples shown in the paper.

```python
import llm_utils
from transformers import AutoModelForCausalLM, AutoTokenizer
import context_steering
import os

os.environ['CUDA_VISIBLE_DEVICES'] = "1"

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf")

query = "Explain Newtonâ€™s second law."

lambda_list = [-3.0, -1.0, 0.0, 1.0, 3.0]
```

```python
context_1 = "I am a toddler. "

base_messages = [{"role": "user", "content": query}]
contextualized_messages = [{"role": "user", "content": context_1 + query}]

print(f"Under context: '{context_1}', the query is '{query}'")
for l in lambda_list:
    print(print(f"Lambda: {l}\n"))
    generation = context_steering.steering_generation(
        base_messages, contextualized_messages, l, model, tokenizer, max_new_length=512
    )
    print(tokenizer.decode(generation, skip_special_tokens=True))
    print("----------------------\n\n")
```
